```python
#BYRJA -- Keyrið til að frumstilla numpy o.fl.
import numpy as np, pandas as pd
import numpy.random as npr
import matplotlib.pyplot as plt
plt.rc('axes', axisbelow=True) # sets zorder=0.5 (default is patch=1,axis=1.5,line=2,text=3)
np.set_printoptions(precision=2, floatmode='fixed', suppress=True)
```

```python
x = np.arange(5)
y = np.sqrt(x)
plt.figure(figsize=(4, 1.8))
plt.plot(x, y, lw=3, ls=':', c='r', marker='o', ms=8);
plt.title('Kvaðratrætur')
plt.xlabel('x'); 
plt.ylabel('y = √x')
```

```python
t = np.linspace(0, 2*np.pi, 30)
s = np.sin(t)
y = npr.random(30)
plt.figure(figsize=(5,2))
plt.plot(t,s,label='sin(x)')
plt.plot(t,y,label='slembitölur')
plt.grid()
plt.legend();
```

```python
x = npr.normal(size = 500)
plt.figure(figsize = (6, 2.5))
plt.hist(x, bins=12, range=(-3,3), rwidth=0.8, color='c', ec='k');
plt.xlabel('x-gildi')
plt.ylabel(f'fjöldi')
plt.grid(axis='y');
      
```

```python
#BYRJA -- Keyrið til að frumstilla numpy o.fl.
import sys
import numpy as np, pandas as pd
import numpy.random as npr
import matplotlib.pyplot as plt
plt.rc('axes', axisbelow=True)
plt.rc('xtick', labelsize = 6)
plt.rc('ytick', labelsize = 6)
%matplotlib inline
np.set_printoptions(precision=2, floatmode='fixed', suppress=True)
```

```python
x = np.linspace(1,6,6)
y = np.sqrt(x)
plt.figure(dpi=144)
```

```python
def f(x,y):
    return np.sin(x)*np.cos(y)

x = np.linspace(0,5)
y = x
z = [[f(xi, yj) for xi in x] for yj in y]
plt.contour(x,y,z,linewidths=3,colors='k')
plt.colorbar()
plt.set_cmap('rainbow')
```
